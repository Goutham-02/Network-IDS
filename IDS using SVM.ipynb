```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
numeric_cols = data.select_dtypes(include=[np.number]).columns
```

```python
std_scaler = StandardScaler()
def normalization(df,col):
  for i in col:
    arr = df[i]
    arr = np.array(arr)
    df[i] = std_scaler.fit_transform(arr.reshape(len(arr),1))
  return df
```

Feature Extraction - HHGS-ROA

```python
from sklearn.ensemble import RandomForestClassifier
import time

class HHGS_ROA:
    def __init__(self, pop_size=30, max_iter=1, num_features=38):
        self.pop_size = pop_size
        self.max_iter = max_iter
        self.num_features = num_features
        self.speed_rate = 0.1
        self.best_solution = None
        self.best_fitness = float('-inf')
        
    def initialize_population(self):
        print("Initializing population...")
        return np.random.randint(2, size=(self.pop_size, self.num_features))
    
    def fitness_function(self, solution, X_train, X_test, y_train, y_test):
        selected_features = np.where(solution == 1)[0]
        if len(selected_features) == 0:
            return 0.0
        
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train[:, selected_features], y_train)
        y_pred = clf.predict(X_test[:, selected_features])
        
        accuracy = np.mean(y_pred == y_test)
        feature_penalty = 0.01 * len(selected_features) / self.num_features
        return accuracy - feature_penalty
    
    def host_feeding(self, solution, best_solution):
        r = np.random.random()
        new_solution = solution + self.speed_rate * r * (best_solution - solution)
        new_solution = np.round(new_solution).astype(int)
        return np.clip(new_solution, 0, 1)
    
    def free_travel(self, solution):
        new_solution = solution.copy()
        mutation_point = np.random.randint(0, self.num_features)
        new_solution[mutation_point] = 1 - new_solution[mutation_point]
        return new_solution
    
    def optimize(self, X_train, X_test, y_train, y_test, columns):
        start_time = time.time()
        print(f"\nStarting HHGS-ROA optimization with {self.pop_size} agents")
        print(f"Dataset shape: Train {X_train.shape}, Test {X_test.shape}")
        
        population = self.initialize_population()
        
        for t in range(self.max_iter):
            iteration_start = time.time()
            old_best = self.best_fitness
            
            for i in range(self.pop_size):
                current_fitness = self.fitness_function(population[i], X_train, X_test, y_train, y_test)
                
                if current_fitness > self.best_fitness:
                    self.best_fitness = current_fitness
                    self.best_solution = population[i].copy()
                    print(f"\nNew best solution found")
                    print(f"Fitness improved from {old_best:.4f} to {self.best_fitness:.4f}")
                    print(f"Number of selected features: {np.sum(self.best_solution)}")
                
                new_solution = self.host_feeding(population[i], self.best_solution) if np.random.random() < 0.5 else self.free_travel(population[i])
                
                new_fitness = self.fitness_function(new_solution, X_train, X_test, y_train, y_test)
                if new_fitness > current_fitness:
                    population[i] = new_solution
            
            if (t + 1) % 10 == 0:
                elapsed = time.time() - iteration_start
                print(f"\nIteration {t+1}/{self.max_iter} completed in {elapsed:.2f}s")
                print(f"Current best fitness: {self.best_fitness:.4f}")
        
        selected_features = np.where(self.best_solution == 1)[0]
        selected_feature_names = [columns[i] for i in selected_features]
        
        total_time = time.time() - start_time
        print("\nOptimization completed!")
        print(f"Total execution time: {total_time:.2f} seconds")
        print(f"Final number of selected features: {len(selected_features)}")
        print(f"Final fitness score: {self.best_fitness:.4f}")
        
        return selected_feature_names, self.best_fitness

# Usage
optimizer = HHGS_ROA(num_features=len(columns))
selected_features, fitness = optimizer.optimize(X_train, X_test, Y_train, Y_test, columns)
print("\nSelected features:", selected_features)
```
